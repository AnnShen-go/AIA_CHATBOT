```python
import ollama

messages = []

def send(chat):
  messages.append(
    {
      'role': 'user',
      'content': chat,
    }
  )
  stream = client.chat(model='yabi/breeze-7b-32k-instruct-v1_0_q8_0', 
    messages=messages,
    stream=True,
  )

  response = ""
  for chunk in stream:
    part = chunk['message']['content']
    print(part, end='', flush=True)
    response = response + part

  messages.append(
    {
      'role': 'assistant',
      'content': response,
    }
  )

  print("")

while True:
    chat = input(">>> ")

    if chat == "/exit":
        break
    elif len(chat) > 0:
        send(chat)
```

## 整理近期簡章

```python
data=[]
import time
base_url="https://aiacademy.tw"
for class_li in class_list[1:]:
  info=[]
  title = re.findall('課程名稱(.*)上課校區',class_li.text)[0]
  print(title)
  info.append(title)# title
  a = class_li.find_all('a', {'title': '簡章'})
  for i in a:
    print(base_url+i['href'])
    info.append(base_url+i['href'])# url
    result = requests.get(url=base_url+i['href'])
    if result.status_code==200:
      soup = BeautifulSoup(result.text, 'html.parser')  # 可以换成 'html.parser' 如果不想用 lxml
      class_info = soup.find('div', {'id': 'bd-l'})
      info.append(class_info.get_text())
    data.append(info)
  time.sleep(1)
print(data)
```

```python
for i in data:
    print(i[2])
```

```python
admission_docs=[]
from importlib import metadata
from langchain.docstore.document import Document # 想要自己處理 string 資料

from langchain.text_splitter import CharacterTextSplitter
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
    separators=["\n\n", "\n","。","，"]
)
for i in data:
  admission=Document(page_content=i[2],metadata={"source":"簡章","url":i[1],"class_name":i[0]})
  # print(admission)
  # print(len(admission.page_content))
  splited_admission = splitter.split_documents([admission])
  admission_docs += splited_admission
for i in admission_docs:
  i.page_content=i.page_content.strip("。，")
  print(len(i.page_content))
#   print((i.page_content))

  
docs += admission_docs
print(len(admission_docs))
```
